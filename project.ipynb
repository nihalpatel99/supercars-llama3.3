```python
from langchain_community.document_loaders import WebBaseLoader

loader = WebBaseLoader("https://en.wikipedia.org/wiki/Porsche_918_Spyder")
page_data = loader.load().pop().page_content
page_data
```

```python
from langchain_core.prompts import PromptTemplate


prompt_extract = PromptTemplate.from_template(
        """
        ### SCRAPED TEXT FROM WEBSITE:
        {page_data}
        ### INSTRUCTION:
        The scraped text is from the car page of wikipedia website.
        Your job is to extract the car specifications and return them in JSON format containing the 
        following keys: `model`, `engine`,'top speed' `chassis` and `modes`.
        Only return the valid JSON.
        ### VALID JSON (NO PREAMBLE):    
        """
)

chain_extract = prompt_extract | llm 
res = chain_extract.invoke(input={'page_data':page_data})
print(res.content)
```

```python
from langchain_core.output_parsers import JsonOutputParser

json_parser = JsonOutputParser()
json_res = json_parser.parse(res.content)
json_res
```

```python
import pandas as pd

df = pd.read_csv("supercars.csv")
df
```

```python
import uuid
import chromadb

client = chromadb.PersistentClient('vectorstore')
collection = client.get_or_create_collection(name="supercars")

if not collection.count():
    for _, row in df.iterrows():
        collection.add(documents=row["model"],
                       metadatas={"model" : row["model"],"engine": row["engine"],"top_speed": row["top_speed"]},
                       ids=[str(uuid.uuid4())])
```

```python
links = collection.query(query_texts=json_res['model'], n_results=2).get('metadatas', [])
links
```

```python
prompt_email = PromptTemplate.from_template(
        """
        ### CAR DESCRIPTION:
        {car_description}
        
        ### INSTRUCTION:
        You are Charles, a business executive sales at Monaco Supercars based in Monaco.
        Your company provides high and ultra rare supercars to VIP customers who you reach out to them via email or phone
        Your job is to write a cold email to the client regarding the cars and their specifications you offer above
        Also add the most relevant ones: {link_list}
        Remeber you are Charlesm business executive sales at Monaco Supercars
        Do not provide a preamble.
        ### EMAIL (NO PREAMBLE):
        
        """
        )

chain_email = prompt_email | llm
res = chain_email.invoke({"car_description": str(json_res), "link_list": links})
print(res.content)
```
